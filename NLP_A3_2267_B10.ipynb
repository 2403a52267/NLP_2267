````python
from collections import Counter
nlp = spacy.load("en_core_web_sm")
text = """Loving the new AI features

 #AI #MachineLearning"""
doc = nlp(text)
nouns = []
verbs = []
for token in doc:
    if token.pos_ in ["NOUN", "PROPN"]:
        nouns.append(token.text)
    elif token.pos_ == "VERB":
        verbs.append(token.text)
noun_freq = Counter(nouns)
verb_freq = Counter(verbs)
print("Noun Frequency:", noun_freq)
print("Verb Frequency:", verb_freq)
````

````python
import json
import nltk
import spacy
from nltk.tokenize import TweetTokenizer
from nltk import pos_tag
from collections import Counter

nltk.download('averaged_perceptron_tagger')
````

````python
file_path = ("/content/enriched_posts.json")
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

type(data), len(data)
````

1)Extract text from posts

````python
texts = []

for item in data:
    if 'text' in item:
        texts.append(item['text'])
    elif 'caption' in item:
        texts.append(item['caption'])
    elif 'content' in item:
        texts.append(item['content'])
texts[:5]
````

2)Tokenize informal text (hashtags, emojis, slang)

````python
# Tokenize informal text (hashtags, emojis, slang)
tokenizer = TweetTokenizer(
    preserve_case=False,
    strip_handles=True,
    reduce_len=True
)

sample_tokens = tokenizer.tokenize(texts[0])
sample_tokens
````

3)POS tagging using NLTK

````python
nltk_pos = pos_tag(sample_tokens)
nltk_pos
````

4)POS tagging using spaCy

````python
nlp = spacy.load("en_core_web_sm")
doc = nlp(texts[0])

[(token.text, token.pos_) for token in doc]
````

````python
print("NLTK POS Tags:")
print(nltk_pos)

print("\nspaCy POS Tags:")
print([(token.text, token.pos_) for token in doc])
````

# *NLTK*

````python
nltk_nouns = [w for w, t in nltk_pos if t.startswith('NN')]
nltk_verbs = [w for w, t in nltk_pos if t.startswith('VB')]
````

# **spaCy**

````python
spacy_nouns = [t.text for t in doc if t.pos_ == 'NOUN']
spacy_verbs = [t.text for t in doc if t.pos_ == 'VERB']
````

````python
print("Top NLTK Nouns:", Counter(nltk_nouns).most_common(10))
print("Top NLTK Verbs:", Counter(nltk_verbs).most_common(10))
````

Discussion

Informal social media text contains emojis, hashtags, slang, and abbreviations
that make POS tagging challenging. NLTK often assigns generic or incorrect tags
to slang and emojis, while spaCy performs better due to contextual embeddings.
However, both tools struggle with noisy text. Hashtags are usually treated as
nouns, helping in topic extraction, while verbs represent opinions and actions.
